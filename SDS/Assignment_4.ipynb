```python
#Task 1

import pandas as pd
from scipy.stats import chi2_contingency

data_homogeneity = pd.DataFrame({
    'City_A_Brand_A': [30, 20],
    'City_A_Brand_B': [40, 30],
    'City_A_Brand_C': [20, 50],
    'City_B_Brand_A': [25, 15],
    'City_B_Brand_B': [35, 25],
    'City_B_Brand_C': [15, 30],
}, index=['City_A', 'City_B'])

chi2_stat, p_value, dof, expected = chi2_contingency(data_homogeneity)

print(f"Chi-square statistic: {chi2_stat}")
print(f"P-value: {p_value}")


alpha = 0.05  # significance level
print(f"\nSignificance level: {alpha}")
print("Result:")
if p_value < alpha:
    print("Reject the null hypothesis. The brand preferences are different between the two cities.")
else:
    print("Fail to reject the null hypothesis. The brand preferences are similar between the two cities.")
```

```python
# Task 3
import scipy.stats as stats


blood_pressure_existing = [120, 122, 118, 124, 123, 121, 125, 119]
blood_pressure_new_drug = [115, 117, 116, 114, 118, 120, 116, 119]

t_statistic, p_value = stats.ttest_ind(blood_pressure_existing, blood_pressure_new_drug)

print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

alpha = 0.05  # significance level
print(f"\nSignificance level: {alpha}")
print("Result:")
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between the blood pressure levels of the existing treatment and the new drug.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the blood pressure levels of the existing treatment and the new drug.")
```

```python
# Task 5
import pandas as pd
from sklearn.linear_model import LinearRegression


data_regression = pd.DataFrame({
    'Study_Hours': [2, 3, 4, 5, 6, 7, 8, 9],
    'Exam_Scores': [50, 55, 60, 65, 70, 75, 80, 85],
})


x = data_regression[['Study_Hours']]
y = data_regression['Exam_Scores']

model = LinearRegression()

model.fit(x, y)

hours_to_predict = [[8]]
predicted_score = model.predict(hours_to_predict)

print(f"Coefficient (slope): {model.coef_[0]}")
print(f"Intercept: {model.intercept_}")
print(f"Predicted Exam Score for 8 hours of study: {predicted_score[0]}")
```
